```python
import vapor
import pandas as pd
```

### Regional PV Pipeline

```python
 opt_vars = ['project_return_aftertax_npv']
#     'marginal_cost_per_kwh'
#     'project_return_aftertax_npv', 
#              'cambium_grid_value', 
#              'project_return_aftertax_npv' 
#              'cambium_co2_rate_avg']

all_scenarios = []

for scenario in ['StdScen19_Low_RE_Cost',
                 'StdScen19_Mid_Case',
                 'StdScen19_High_RE_Cost']:

    for tech in ['pv','wind']:
        
        for batt_size in [0, 25]:
            
            vapor.config.CAMBIUM_SCENARIO = scenario
            # --- Initialize Pipeline ---
            regional_plumbing = vapor.RegionalPipeline(
                         tech=tech,
                         aggregate_region='state',
                         aggregate_func='sum',
                         re_capacity_mw=100,
                         batt_capacity_mw=(batt_size*.99, batt_size))

            # --- Execute Simulations ---
            regional_plumbing.setup()

            regional_plumbing.run(opt_vars)

            # --- Get Best Systems ---
            regional_best = regional_plumbing.best_systems

            # --- Save ---
            regional_best['scenario'] = scenario
            regional_best['batt_size'] = batt_size
            regional_best.to_pickle(f'state_best_{tech}_{scenario}_batt{batt_size}.pkl')
            all_scenarios.append(regional_best)

results_df = pd.concat(all_scenarios, axis='rows')
```

```python
import pandas as pd

# results_df['marginal_cost_mwh'] *= 1000

# --- Load Geometry ---
regional_geography = pd.read_pickle('regional_geography.pkl')

Viz = vapor.Visualizer(results=results_df,
                       geo=regional_geography,
                       region_label_pct=0.2)

# --- Choropleths ---
Viz.choropleth(column='marginal_cost_mwh', scenario='StScen19_Mid_Case',
               ascending=True, reverse_cmap=True,
              title="Lifetime Marginal Cost of RE ($/MWh)")
Viz.choropleth(column='lifetime_cambium_total_value', scenario='StScen19_Mid_Case',
               ascending=False,
              title="Lifetime Cambium Total Grid Value ($ Mil.)")
Viz.choropleth(column='lifetime_co2_rate_marginal', scenario='StScen19_Mid_Case',
               ascending=False,
              title="Lifetime Marginal CO2 Savings (Mil. Lbs)")

```
