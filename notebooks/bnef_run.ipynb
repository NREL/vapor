```python
import vapor
```

### Regional PV Pipeline

```python
 opt_vars = ['project_return_aftertax_npv']
#     'marginal_cost_per_kwh'
#     'project_return_aftertax_npv', 
#              'cambium_grid_value', 
#              'project_return_aftertax_npv' 
#              'cambium_co2_rate_avg']
    
for tech in ['wind','pv']:
    # --- Initialize Pipeline ---
    regional_plumbing = vapor.RegionalPipeline(
                 tech=tech,
                 aggregate_region='state',
                 aggregate_func='sum',
                 re_capacity_mw=100,
                 batt_capacity_mw=(0,100))

    # --- Execute Simulations ---
    regional_plumbing.setup()

    regional_plumbing.run(opt_vars)

    # --- Get Best Systems ---
    regional_best = regional_plumbing.best_systems
    
    # --- Save ---
    regional_best.to_pickle(f'regional_best_{tech}.pkl')
```

```python
# --- Choropleths ---
Viz = vapor.Visualizer(
                    tech='wind',
#                     aggregate_region='state',
                    results=regional_best,
                    geo=regional_plumbing.geometry)
Viz.choropleth(column='project_return_aftertax_npv')
```
